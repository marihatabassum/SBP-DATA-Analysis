```python
sbp_csv=pd.read_csv("/content/sbp_data.csv")
sbp_csv
```

```python
sbp_csv['geometry'] =  sbp_csv['geometry'].apply(lambda x: x.replace('(','').replace(')',''))
sbp_csv
```

```python
sbp_csv[["Point", "NO", "X", "Y", "Z"]]=sbp_csv.geometry.str.split(" ", expand=True)
sbp_csv
```

```python
sbp_csv.to_csv("sbp_xyz.csv", sep=",", index=True, encoding="utf-8")
sbp_xyz=pd.read_csv("/content/sbp_xyz.csv")
sbp_xyz
```
